```python
# List of sectors to keep
sectors_to_keep = ['Tourism and Recreation', 'Ocean Economy', 'Living Resources', 'Offshore Mineral Resources', 'Ship and Boat Building']

df = df[df['sector'].isin(sectors_to_keep)]
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt

df.drop('Unnamed: 0', axis=1, inplace=True)

# Define predictors and target variables
predictors = ['adt', 'sla', 'ugos', 'ugosa', 'vgos', 'vgosa', 'establishments', 'employment']
target_variables = 'gdp'

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df[predictors], df[target_variables], test_size=0.2, random_state=42)
```

```python
# Define AdaBoost regressor
ada = AdaBoostRegressor()

# Define the parameter grid for hyperparameter tuning
param_grid = {
    'n_estimators': [50, 100, 150],  # Number of estimators (weak learners)
    'learning_rate': [0.01, 0.1, 1.0]  # Learning rate
}

grid_search = GridSearchCV(estimator=ada, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5)
grid_search.fit(X_train, y_train)


grid_search.fit(X_train, y_train)

y_pred = grid_search.predict(X_test)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
para_grid_rf = {'n_estimators': [50, 100, 150],
                 'max_depth': [None, 10, 20],
                 'min_samples_split': [2, 5, 10],
                 'min_samples_leaf': [1, 2, 4]}
para_grid_gb = {'n_estimators': [50, 100, 150],
                 'learning_rate': [0.01, 0.1, 1.0],
                 'max_depth': [3, 5, 7]}
para_grid_ab = {'n_estimators': [50, 100, 150],
                 'learning_rate': [0.01, 0.1, 1.0]}
para_grid_dt = {'max_depth': [None, 10, 20],
                 'min_samples_split': [2, 5, 10],
                 'min_samples_leaf': [1, 2, 4]}

models = {
    'Random Forest': (RandomForestRegressor(), para_grid_rf),
    'Gradient Boosting': (GradientBoostingRegressor(), para_grid_gb),
    'AdaBoost': (AdaBoostRegressor(), para_grid_ab),
    'Decision Tree': (DecisionTreeRegressor(), para_grid_dt)
}

results = {}
for name, (model, para_grid) in models.items():
    md = GridSearchCV(model, para_grid, scoring='neg_mean_squared_error', cv=5)
    md.fit(X_train, y_train)
    pred = md.predict(X_test)
    mse = mean_squared_error(y_test, pred)
    results[name] = mse

for name, mse in results.items():
    print(f"{name}: mse - {mse}")
```
